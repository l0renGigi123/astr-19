```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Generate 1000 random numbers btwn 0 and 1
data = np.random.uniform(0, 1, 1000)
```

```python
#histogram with 100 bins
plt.figure(figsize=(10, 6))
plt.hist(data, bins=100, color='blue', edgecolor='black', alpha=0.7)

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Uniformly Distributed Random Numbers')

#save/show figure
plt.savefig("Prompt9.pdf",bbox_inches="tight", facecolor = "white") 
plt.show()
```
