```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate 1000 nums between 1 and 0
data = np.random.poisson(lam=5.0, size=1000)
```

```python
#histogram with 100 bins
plt.figure(figsize=(10, 6))
plt.hist(data, bins=100, color='blue', edgecolor='black', alpha=0.7)

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Uniformly Distributed Random Numbers')

#save/show figure
plt.savefig("Prompt10.pdf",bbox_inches="tight", facecolor = "white") 
plt.show()
```
